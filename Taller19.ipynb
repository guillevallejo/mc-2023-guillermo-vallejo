```python
import numpy as np
import matplotlib.pyplot as plt

def linear_regression(x, y):
    A = np.vstack([x, np.ones(len(x))]).T
    m, b = np.linalg.lstsq(A, y, rcond=None)[0]
    return m, b

def calculate_statistics(x, y):
    n = len(x)
    m, b = linear_regression(x, y)

    # Cálculo de los valores ajustados
    y_pred = m * x + b

    # Cálculo de la suma de residuos al cuadrado
    residuals = y - y_pred
    sum_squared_residuals = np.sum(residuals ** 2)

    # Cálculo de la desviación estándar (Sy)
    std_deviation = np.sqrt(sum_squared_residuals / (n - 2))

    # Cálculo del error estándar de la estimación (s y/x)
    std_error_estimation = std_deviation / np.sqrt(np.sum((x - np.mean(x)) ** 2))

    # Cálculo del coeficiente de correlación (r)
    correlation_matrix = np.corrcoef(x, y)
    correlation_coefficient = correlation_matrix[0, 1]

    return std_deviation, std_error_estimation, correlation_coefficient

x = np.array([1, 2, 3, 4, 5, 6, 7])
y = np.array([0.2, 0.5, 1.8, 3.4, 5.7, 9.0, 13.8])

# Regresión lineal
m, b = linear_regression(x, y)

# Cálculo de las estadísticas
std_deviation, std_error_estimation, correlation_coefficient = calculate_statistics(x, y)

# Gráfico de dispersión de puntos
plt.scatter(x, y)

# Regresión lineal
x_fit = np.linspace(x[0], x[-1], 100)
y_fit = m * x_fit + b
plt.plot(x_fit, y_fit, 'r', label='Lineal')

plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.title('Regresión lineal con estadísticas')
plt.legend()
plt.show()

print("Desviación estándar (Sy): {:.5f}".format(std_deviation))
print("Error estándar de la estimación (s y/x): {:.5f}".format(std_error_estimation))
print("Coeficiente de correlación (r): {:.5f}".format(correlation_coefficient))
```
