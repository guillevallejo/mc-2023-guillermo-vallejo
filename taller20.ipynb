```python
import numpy as np
import matplotlib.pyplot as plt

def quadratic_regression(x, y):
    A = np.vstack([x**2, x, np.ones(len(x))]).T
    a, b, c = np.linalg.lstsq(A, y, rcond=None)[0]
    return a, b, c

def calculate_correlation_coefficient(x, y):
    correlation_matrix = np.corrcoef(x, y)
    correlation_coefficient = correlation_matrix[0, 1]
    return correlation_coefficient

x = np.array([0, 1, 2, 3, 4, 5, 6])
y = np.array([4.2, 1.4, 0, -0.4, -0.1, 1.6, 4.1])

# Ajuste del polinomio de segundo grado
a, b, c = quadratic_regression(x, y)

# Cálculo del coeficiente de correlación
correlation_coefficient = calculate_correlation_coefficient(x, y)

# Gráfico de dispersión de puntos
plt.scatter(x, y)

# Polinomio de segundo grado
x_fit = np.linspace(x[0], x[-1], 100)
y_fit = a * x_fit**2 + b * x_fit + c
plt.plot(x_fit, y_fit, 'r', label='Polinomio de segundo grado')

plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.title('Ajuste polinomial de segundo grado')
plt.legend()
plt.show()

print("Coeficiente de correlación (r): {:.5f}".format(correlation_coefficient))
```
