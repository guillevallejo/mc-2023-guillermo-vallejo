```python
# Punto 2
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

x = np.array([1, 2, 3, 4, 5, 6, 7, 8])
y = np.array([3.8, 4, 4.2, 4.6, 4.9, 5.3, 5.7, 6.1])

# Función exponencial: y = a * exp(b * x)
def exponential_func(x, a, b):
    return a * np.exp(b * x)

# Función de potencias: y = a * x^b
def power_func(x, a, b):
    return a * np.power(x, b)

# Función de crecimiento lineal: y = a + b * x
def linear_growth_func(x, a, b):
    return a + b * x

# Ajuste exponencial
exp_params, _ = curve_fit(exponential_func, x, y)

# Ajuste de potencias
power_params, _ = curve_fit(power_func, x, y)

# Ajuste de crecimiento lineal
linear_params, _ = curve_fit(linear_growth_func, x, y)

x_pred = np.linspace(1, 8, 100)

y_exp_pred = exponential_func(x_pred, *exp_params)
y_power_pred = power_func(x_pred, *power_params)
y_linear_pred = linear_growth_func(x_pred, *linear_params)

# Gráfica 1: Exponencial
plt.figure(figsize=(8, 6))
plt.scatter(x, y, label='Puntos')
plt.plot(x_pred, y_exp_pred, label='Exponencial')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ajuste exponencial (Mejor Aproximación)')
plt.grid(True)
plt.show()

# Gráfica 2: Potencias
plt.figure(figsize=(8, 6))
plt.scatter(x, y, label='Puntos')
plt.plot(x_pred, y_power_pred, label='Potencias')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ajuste de potencias')
plt.grid(True)
plt.show()

# Gráfica 3: Crecimiento lineal
plt.figure(figsize=(8, 6))
plt.scatter(x, y, label='Puntos')
plt.plot(x_pred, y_linear_pred, label='Crecimiento lineal')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ajuste de crecimiento lineal')
plt.grid(True)
plt.show()
```

```python
# Punto 3
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 3, 5, 7, 9])
y = np.array([2.8, 3, 3.5, 3.8, 4.5])

def lagrange_interpolation(x, y, degree, x_interp):
    n = degree + 1  # Número de puntos de interpolación
    m = len(x)  # Número de puntos conocidos
    
    # Verificar que el grado de interpolación no exceda el número de puntos conocidos
    if n > m:
        raise ValueError("El grado de interpolación no puede ser mayor que el número de puntos conocidos.")
    
    # Calcular los polinomios de Lagrange
    lagrange_polynomials = []
    for i in range(n):
        lagrange_poly = 1
        for j in range(n):
            if j != i:
                lagrange_poly *= (x_interp - x[j]) / (x[i] - x[j])
        lagrange_polynomials.append(lagrange_poly)
    
    # Calcular el valor interpolado
    interp_value = 0
    for i in range(n):
        interp_value += y[i] * lagrange_polynomials[i]
    
    return interp_value

x_interp = 4.75

# Grado 1
interp_value_1 = lagrange_interpolation(x, y, 1, x_interp)
y_poly_1 = lagrange_interpolation(x, y, 1, x)

# Grado 2
interp_value_2 = lagrange_interpolation(x, y, 2, x_interp)
y_poly_2 = lagrange_interpolation(x, y, 2, x)

# Grado 3
interp_value_3 = lagrange_interpolation(x, y, 3, x_interp)
y_poly_3 = lagrange_interpolation(x, y, 3, x)

# Grado 4
interp_value_4 = lagrange_interpolation(x, y, 4, x_interp)
y_poly_4 = lagrange_interpolation(x, y, 4, x)

plt.figure(figsize=(8, 6))

# Gráfica 1: Grado 1
plt.subplot(221)
plt.scatter(x, y, color='red', label='Puntos conocidos')
plt.plot(x, y_poly_1, label='Grado 1')
plt.scatter(x_interp, interp_value_1, color='blue', marker='x', label='f(4.75) Grado 1')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Polinomio de interpolación - Grado 1')
plt.grid(True)

# Gráfica 2: Grado 2
plt.subplot(222)
plt.scatter(x, y, color='red', label='Puntos conocidos')
plt.plot(x, y_poly_2, label='Grado 2')
plt.scatter(x_interp, interp_value_2, color='green', marker='x', label='f(4.75) Grado 2')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Polinomio de interpolación - Grado 2')
plt.grid(True)

# Gráfica 3: Grado 3
plt.subplot(223)
plt.scatter(x, y, color='red', label='Puntos conocidos')
plt.plot(x, y_poly_3, label='Grado 3')
plt.scatter(x_interp, interp_value_3, color='purple', marker='x', label='f(4.75) Grado 3')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Polinomio de interpolación - Grado 3')
plt.grid(True)

# Gráfica 4: Grado 4
plt.subplot(224)
plt.scatter(x, y, color='red', label='Puntos conocidos')
plt.plot(x, y_poly_4, label='Grado 4')
plt.scatter(x_interp, interp_value_4, color='orange', marker='x', label='f(4.75) Grado 4')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Polinomio de interpolación - Grado 4')
plt.grid(True)

plt.tight_layout()
plt.show()
```
